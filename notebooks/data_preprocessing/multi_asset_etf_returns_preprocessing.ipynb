```python
# Libraries
import pandas as pd
```

```python
# Load raw dataset
# Raw files stay untouched inside /raw

df = pd.read_csv("../../data/raw/multi_asset_etf_returns.csv")

print("Raw Dataset:")
print(df.head())
```

```python


# Convert month-year strings into datetime format
df["Mes"] = pd.to_datetime(df["Mes"], format="%y-%b")

# Rename column for professionalism
df.rename(columns={"Mes": "Date"}, inplace=True)
```

```python
# Monthly returns are currently stored as strings
# Example: "3.15%"
# Convert them into numeric decimal returns
# Identify return columns

return_columns = [
    "Treasury ETF",
    "Technology ETF",
    "Gold",
    "Global Bank ETF",
    "E MER ETF"
]
```

```python
# Remove % and convert to decimal
for col in return_columns:
    df[col] = (
        df[col]
        .str.replace("%", "", regex=False)
        .astype(float) / 100
    )
```

```python
# Check for missing observations
# Missing values can affect:
# volatility estimation
# correlation structure
# portfolio metrics

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Validate data types after preprocessing

print("\nData Types:")
print(df.dtypes)
```

```python
# Quick validation checks

print("\nDataset Shape:")
print(df.shape)

print("\nSummary Statistics:")
print(df.describe())
```

```python
# Export clean dataset
# This dataset will be used in downstream analysis notebooks

output_path = "../../data/processed/pri_multi_asset_etf_returns.csv"

df.to_csv(output_path, index=False)

print(f"\nClean dataset exported to:\n{output_path}")
```
